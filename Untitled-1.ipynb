```python
import numpy as np
import pandas as pd
import keras
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
vec = keras.layers.TextVectorization()
vec.adapt(pd.concat([train['name1'], test['name1']]) )
```

```python
vectorized_train = vec(train['name1']).numpy()
vectorized_train.shape
```

```python
vec.vocabulary_size()
```

```python
train['cat_id']
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(vectorized_train, train['cat_id'])
```

```python
submission = pd.DataFrame()
soft = lg.predict(vec(test['name1']))
submission['cat_id'] = soft
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(vectorized_train, train['cat_id'])
```

```python
submission = pd.DataFrame()
soft = knn.predict(vec(test['name1']))
submission['cat_id'] = soft
```
